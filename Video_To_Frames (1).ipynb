```python
import cv2
import numpy as np
import os

# set video file path of input video with name and extension
vid = cv2.VideoCapture('/content/drive/MyDrive/Vid2Frames/Test_Video_0-10.mp4')
vid.get(20)


#if not os.path.exists('images'):
#    os.makedirs('images')

#for frame identity
index = 0
while(True):
    # Extract images
    ret, frame = vid.read()
    # end of frames
    if not ret: 
        break
    # Saves images
    name = '/content/drive/MyDrive/Vid2Frames/0-10/frame' + str(index) + '.jpg'
    print ('Creating...' + name)
    cv2.imwrite(name, frame)

    # next frame
    index += 1
```
